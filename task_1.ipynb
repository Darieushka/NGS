```python
from Bio import SeqIO
import matplotlib.pyplot as plt
import time
import seaborn as sns

le=100
a= [0] * 100
b= [0] * 100
a2 = [0] * 100
b2 = [0] * 100
a3 = [0] * 100
b3 = [0] * 100
c1 = 0
c2 = [0.001] * 100
c3 = 0.001


for record in SeqIO.parse("100x.1.fastq", "fastq"): 
    c = record.seq
    d = record.letter_annotations["phred_quality"]
    c1 += 1
    for n in range(0, le):
            if c[n] in {'G','C'}:
                a[n] += 1
            if d[n] >= 15:
                c2[n] += 1
                if c[n] in {'G','C'}:
                    a2[n] += 1
    if sum(i > 30 for i in d)>80:
        c3+=1
        for n in range(0,le):
            if c[n] in {'G','C'}:
                a3[n]+=1
        

for n in range(0, le):
    b[n]=a[n]/c1
    b2[n]=a2[n]/c2[n]
    b3[n]=a3[n]/c3

    
plt.figure()
sns.kdeplot(b)   
plt.figure()
sns.kdeplot(b2)
plt.figure()
sns.kdeplot(b3)
```
