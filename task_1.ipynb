```python
from Bio import SeqIO
import matplotlib.pyplot as plt
import time
import seaborn as sns

le=100

u1=[]
u2=[]
u3=[]

for record in SeqIO.parse("100x.1.fastq", "fastq"): 
    
    c = record.seq
    d = record.letter_annotations["phred_quality"]
 
    u1.append((c.count('C')+c.count('G'))/100)
    
    a2=0
    c2=0.001
    for n in range(0, le):
            if d[n] >= 20:
                c2 += 1
                if c[n] in {'G','C'}:
                    a2 += 1
    u2.append(a2/c2)
    
    if sum(i > 30 for i in d)>80:
        u3.append((c.count('C')+c.count('G'))/100)
        

plt.figure()
sns.distplot(u1)   
plt.figure()
sns.distplot(u2)
plt.figure()
sns.distplot(u3)
```
